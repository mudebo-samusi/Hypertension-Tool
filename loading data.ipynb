```python
import pandas as pd

# Load the dataset
df = pd.read_csv("simulated_bp_data.csv")

# Display the first 5 rows
print(df.head())
```

```python
import matplotlib.pyplot as plt
```

```python
# Filter data for a single patient
patient_1_data = df[df["Patient ID"] == 1]
print(patient_1_data)
```

```python
print(patient_1_data.head())  # Check if data exists
print(f"Total readings for Patient 1: {len(patient_1_data)}")
```

```python
# Plot Systolic and Diastolic BP over time
plt.figure(figsize=(10, 6))
plt.plot(patient_1_data["Timestamp"], patient_1_data["Systolic BP"], label="Systolic BP")
plt.plot(patient_1_data["Timestamp"], patient_1_data["Diastolic BP"], label="Diastolic BP")
plt.xlabel("Time")
plt.ylabel("Blood Pressure (mmHg)")
plt.title("Blood Pressure Trends for Patient 1")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
